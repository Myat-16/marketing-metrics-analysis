### Cohort Analysis

What Cohort Analysis Does
Cohort analysis tracks groups of customers over time to understand:

Customer retention - How many customers come back each month?
Customer lifetime value - How much revenue do different groups generate?
Acquisition quality - Which months bring the best long-term customers?

```python
print(df['InvoiceDate'].min(), df['InvoiceDate'].max())
```

```python
df['Year'] = df['InvoiceDate'].dt.year
df['Month'] = df['InvoiceDate'].dt.month
```

```python
df['InvoiceMonth'] = df['Year'].astype(str) + '-' + df['Month'].astype(str) + '-01'
```

We want their first ever purchase month to determine their cohort

```python
pivot_table = pd.pivot_table(df, values='InvoiceMonth', index='CustomerID', aggfunc=np.min)
print(pivot_table)
```

Pareto Analysis - Finding High values Customer(Customer & Revenue)

```python
pareto_t_c = pd.pivot_table(df, values='Revenue', index='CustomerID', aggfunc=np.sum).sort_values(by='Revenue', ascending=False)
print(pareto_t_c)
```

```python
pareto_t_c['percenttotal'] = pareto_t_c.divide(pareto_t_c.sum())
```

```python
pareto_t_c
```

```python
pareto_t_c['cumulative'] = pareto_t_c['percenttotal'].cumsum()
pareto_t_c
```

```python
pareto_t_c = pareto_t_c[pareto_t_c['cumulative'] < 0.8].sort_values(by='Revenue', ascending=False).reset_index()
```

```python
pareto_t_c['CustomerID'] = pareto_t_c['CustomerID'].astype(str)
```

```python
# Create the primary plot
fig, ax1 = plt.subplots()
sns.barplot(x=pareto_t_c['CustomerID'], y=pareto_t_c['Revenue'], ax=ax1, color='blue')
ax1.set_ylabel('Revenue', color='blue')

# Create the secondary axis
ax2 = ax1.twinx()
sns.lineplot(x=pareto_t_c['CustomerID'], y=pareto_t_c['cumulative'], ax=ax2, color='red', label='cumulative')
ax2.set_ylabel('Cumulative', color='red')
ax2.tick_params(axis='y', labelcolor='red')
```

RFM analysis


```python
frequency = pd.pivot_table(df, values='InvoiceNo', index='CustomerID', aggfunc= 'nunique')
frequency
```

```python
frequency.columns = ['FrequencyRate']
frequency
```

```python
frequency.sort_values(by='FrequencyRate', ascending=False)
```

```python
frequency ['FrequencyScore'] = pd.qcut(frequency['FrequencyRate'].rank(method = 'first'), q=5, labels=[1,2,3,4,5])
```

```python
frequency
```

```python
df['InvoiceDate'].max()
```

```python
recency = pd.pivot_table(df, values='InvoiceDate', index='CustomerID', aggfunc= 'max')
```

```python
recency
```

```python
recency['day_diff'] = (df['InvoiceDate'].max() - recency['InvoiceDate']).dt.days
recency
```

```python
recency ['recencyScore'] = pd.qcut(recency['day_diff'].rank(method = 'first'), q=5, labels=[5,4,3,2,1])
```

```python
recency
```

```python
monetary = pd.pivot_table(df, values='Revenue', index='CustomerID', aggfunc= 'sum')
monetary
```

```python
monetary ['monetaryScore'] = pd.qcut(monetary['Revenue'].rank(method = 'first'), q=5, labels=[1,2,3,4,5])
```

```python
monetary
```

```python
rfm_t = pd.merge (frequency[['FrequencyScore']], recency[['recencyScore']], on='CustomerID')
rfm_t
```

```python
rfm_t = pd.merge (rfm_t, monetary[['monetaryScore']], on='CustomerID')
rfm_t
```

```python
segments = []
for fs,rs in zip (rfm_t['FrequencyScore'],rfm_t['recencyScore']):
  if fs >= 4 and rs >= 5:
    segments.append("champions")
  elif fs >= 4 and rs >=3 and rs <=4:
    segments.append("loyal")
  elif fs >= 5 and rs<= 2:
    segments.append("cannot lose them")
  elif fs >= 3 and fs <= 4 and rs <= 2:
    segments.append("at risk")
  elif fs == 3 and rs == 3:
    segments.append("need attention")
  elif fs <= 2 and rs == 3:
    segments.append("about to sleep")
  elif fs >= 2 and fs <= 3 and rs >= 4:
    segments.append("potential loyalists")
  elif fs == 1 and rs == 4:
    segments.append("promising")
  elif fs == 1 and rs == 5:
    segments.append("new customers")
  else:
    segments.append("hibernating")
```

```python
rfm_t['segmentations'] = segments
```
